```python
target = df['Touchscreen']

dummie_col = ['Brand',
              'Model Name',
              'Dial Shape',
              'Strap Color',
             'Strap Material',
             'Bluetooth',
             'Display Size',
             'Weight',
           
] 
```

```python
df = pd.get_dummies(df, columns= dummie_col)
```

```python
drop_col = ['Touchscreen',
            'Current Price',
            'Battery Life (Days)'
    
]
df = df.drop(drop_col, axis=1)
```

```python
df.head().T
```

```python
scaling_columns = ['Original Price','Number OF Ratings','Discount Percentage'
    
]

for column in scaling_columns:
    df[column] = (df[column] - df[column].mean())/df[column].std()
df.head().T
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df, target,
    test_size=0.3, shuffle=True, random_state=42, 
)
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import cross_val_score, KFold
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
models ={'Logistic Regression': LogisticRegression(),
        'KNN': KNeighborsClassifier(n_neighbors=3),
        'Decision Tree Classifier': DecisionTreeClassifier()}
```

```python
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_scaled, y_train)
logreg.score(X_train_scaled,y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled,y_train)
knn.score(X_train_scaled,y_train)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train_scaled,y_train)
dt.score(X_train_scaled,y_train)
```

```python
# Make predictions
y_preds = dt.predict(X_test_scaled)
y_preds[:5]
```

```python
# Try different numbers of estimators (trees)... (no cross-validation)
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    model = DecisionTreeClassifier().fit(X_train_scaled, y_train)
    print(f"Model accuracy on test set: {model.score(X_test_scaled, y_test) * 100}%")
    print("")
```

```python
from sklearn.model_selection import cross_val_score

# With cross-validation
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    dt.fit(X_train_scaled, y_train)
    print(f"Model accuracy on test set: {dt.score(X_test_scaled, y_test) * 100}%")
    print(f"Cross-validation score: {np.mean(cross_val_score(dt, X_train_scaled,y_train, cv=5)) * 100}%")
    print("")
```

```python
# Accuracy as percentage
print(f"Fitness Trackers Products Ecommerce Classifier Accuracy: {dt.score(X_test, y_test) * 100:.2f}%")
```
