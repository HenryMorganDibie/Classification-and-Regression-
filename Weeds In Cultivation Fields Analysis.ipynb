```python
df['binome.discoweed'] = df['binome.discoweed'].fillna('0')
df['family'] = df['family'].fillna('0')

df['taxo'] = df['taxo'].fillna(0)
df['binome.discoweed.noinfra'] = df['binome.discoweed.noinfra'].fillna(0)
df["taxref10.CD_REF.infra"] = df["taxref10.CD_REF.infra"].fillna(0)
df["lifeform"] = df["lifeform"].fillna(0)
df["lifeform"] = df["lifeform"].fillna(0)
df["biogeo"] = df["biogeo"].fillna(0)

df["c"] = df["c"].fillna(0)
df["z"] = df["z"].fillna(0)
df["module"] = df["module"].fillna(0)
df["taxref10.CD_REF"] = df["taxref10.CD_REF"].fillna(0)
```

```python
df.isnull().sum()
```

```python
FR = df['red.FR'].value_counts()
print(FR)
FR.plot.barh(y = FR.values, figsize = (8,5))
```

```python
df['red.FR'].replace(['VU', 'LC', 'DD', 'EN', 'CR', 'NT', 'RE', 'EW', 'EX'],
                        [0, 1, 2, 3, 4, 5, 6, 7, 8], inplace=True)
```
